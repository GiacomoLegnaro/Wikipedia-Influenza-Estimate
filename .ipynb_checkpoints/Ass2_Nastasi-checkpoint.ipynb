The aim of this assignment is to predict when there will be the peak of Influenza epidemic analyzing the pageviews of some words on Wikipedia.  
  
  
The ground truth is the Influnet incidence. Influnet is the national system of epidemiological and virological surveillance, coordinated by the Ministero della Salute, which relies on the collaboration of the Istituto Superiore di Sanità (ISS), the Centro Interuniversitario per la Ricerca sull’Influenza (CIRI), doctors and paediatricians, the network of peripheral laboratories for influenza and regional Departments of Health.  
The epidemiological surveillance objectives are to estimate the weekly incidence of influenza illness during the winter season, in order to measure the duration and intensity of the outbreak.  
The objectives of the virological surveillance are: monitoring the movement of different types and subtypes of influenza viruses in different geographical areas and in different periods of the epidemic season, assess the antigenic homology between epidemic strains and vaccine strains, through serological and molecular analysis of clinical samples taken from patients with flu symptoms.
  
The value of the ground truth is the fraction in the population f new weekly cases, so try to estimate that with Wikipedia pageviews is a good idea because people looking on the internet to inform themself when they start to get sick.

### Process the Wikipedia pageview raw data

```python
# function that read the .txt file of ground truth and select just the column of our interest (Settimana, incidenza_tot)
def read_groud_truth(directory_file):
    # from .txt to dataFrame
    df = pd.read_table(directory_file, sep=' ', names=['Settimana', 'Tot_Medici', 'Tot_casi', 'Tot_assistiti', 
                            'incidenza_tot', '', '', '', '', '', '', '', ''], header=0, quoting=csv.QUOTE_NONE)
    df = df[['Settimana','incidenza_tot']] # select columnss
    # from dataFrame to dictionary of type {'Settimana': incidenza_tot}
    dicti = df.set_index('Settimana')['incidenza_tot'].to_dict() 
    for w,num in dicti.items():
        dicti[w] = float(num.replace(',','.')) # in the ground truth txt file the decimal separator is a comma instead of dot
    return dicti
```

```python
############ DOWNLOAD #######
url = 'https://dumps.wikimedia.org/other/pagecounts-raw/' # general link of Page view statistics for Wikimedia projects
if not os.path.exists('./raw_data'): # creation of folder where save the raw data in format .gz
    os.mkdir('./raw_data')
    
year_list = ['2014', '2015', '2016']
gt_date = gt_14_15_dict.keys()+gt_15_16_dict.keys()
    
for year in year_list: 
    url2 = 'https://dumps.wikimedia.org/other/pagecounts-raw/'+year # year link
    for month in ['-01', '-02', '-03', '-04', '-10', '-11', '-12']: # month to analyze (just the month contained in the GroundTruth)
        links = [] # list of file to download
        html_page = urllib2.urlopen(url2+'/'+year+month) # year-month link
        soup = BeautifulSoup(html_page)
        for link in soup.findAll('a', attrs={'href': re.compile("^pagecounts")}): 
            # take all the links starting with 'pagecounts' (which are the .gz file)
            link = link.get('href')
            # extract week from the date
            week = datetime.strptime(os.path.splitext(link)[0].split('-')[1], '%Y%m%d').isocalendar()[1]
            if year+'-'+"{0:0=2d}".format(week) in set(gt_date):
                links.append(link)
        for index in links:
            print 'Downloading file '+ index
            file_url = url2+'/'+year+month+'/'+index # link of the single file
            wget.download(file_url, './raw_data/') # download .gz file
```

### Aggregation of the pageviews on a weekly time scale
The raw data are in the format YearMonthDay-HourMinuteSecond so it needs to aggregate (summing) the data of the same week

```python
# function that sum all the values of the same week

# INPUTS:
# - words_list = list of words to choose from all it pages
# - selection_hour = if True --> consider just click during the day (exclude click in the hour of night)
#                    if False --> consider all hours

def words_dictionary(words_list, selection_hour):
    seasons_dict = collections.OrderedDict() # ordered dict so the it is maintained the temporal sequence
    list_seas = [i for i in sorted(os.listdir('./it_wikipedia/'))] # list of seasons
    for season in list_seas:
        print 'SEASON: '+season
        seasons_dict[season] = collections.OrderedDict() # {'2014-2015: {}, '2015-2016': {}}
        week_list = [j for j in sorted(os.listdir('./it_wikipedia/'+season))] # list of year-week in the current season
        for week in week_list:
            print '\t week: '+ week
            seasons_dict[season][week] = collections.OrderedDict() 
                  # ex {'2014-2015': {'2014-42': {}, ..., '2015-17': {}}, '2015-2016'...}
            if selection_hour == True: # exclude hour from 00 to 05
                data_list = [k for k in sorted(os.listdir('./it_wikipedia/'+season+'/'+week)) 
                             if datetime.strptime(os.path.splitext(k)[0].split('-')[2], '%H%M%S').hour >= 6 
                             and datetime.strptime(os.path.splitext(k)[0].split('-')[2], '%H%M%S').hour <= 23] 
            else: # include all hours
                data_list = [k for k in sorted(os.listdir('./it_wikipedia/'+season+'/'+week))]
            for data in data_list:
                # from .txt file to dataFrame
                temp_df = pd.read_table('./it_wikipedia/'+season+'/'+week+'/'+data, sep=' ', 
                                        names=['domain_code', 'page_title', 'count_views', 'total_response_size'], 
                                        quoting=csv.QUOTE_NONE)
                # select rows of words in words_list (words of our interest)
                temp_df = temp_df.loc[temp_df['page_title'].isin(words_list)] 
                temp_dict = temp_df.set_index('page_title')['count_views'].to_dict() # from dataFrame to dictionary
                # sum hour value
                seasons_dict[season][week] = dict(Counter(seasons_dict[season][week])+Counter(temp_dict))
            # ex {'2014-2015': {'2014-42': {'Influenza': tot click in week 2014-42}}, '2015-2016': ...}
    return seasons_dict
```

```python
# function that create a matrix array having as column the values of serie of words in list features and list_features_trend
# if prec_week == True --> a columns are added, these columns are the same words in list features but with values of 
#.        preceding week, and the column of Influnet incidence for the week preceding the target week
 
def training_data(list_features, list_features_trend, prec_week):
    train = np.array(words_values_hours[list_features[0]]).reshape(-1,1) # create matrix with the first serie of values  
    for feat in list_features[1:]: # from the second word because the first has been already added
        train = np.column_stack((train, words_values_hours[feat])) # add the new serie values in column
    for trend in list_features_trend:
        train = np.column_stack((train, words_values_trend[trend]))
    if prec_week == False:
        return train
    else: # add preceding week
        train = np.column_stack((train[1:,:], train[:-1,:], gt_all[:-1]))
        return train
```

The approach used to find a good model to predict the Influnet incidence is doing feature selection in a different way, select a certain number od features and then build the linear regression.  
  
The feature selection methods used are:
* $LinearRegression$: ordinary least squares Linear Regression
* $Ridge$: linear least squares with l2 regularization. This model solves a regression model where the loss function is the linear least squares function and regularization is given by the l2-norm
* $Lasso$ : linear model trained with L1 prior as regularizer (aka the Lasso)
* $RandomizedLasso$: randomized Lasso works by subsampling the training data and computing a Lasso estimate where the penalty of a random subset of coefficients has been scaled. By performing this double randomization several times, the method assigns high scores to features that are repeatedly selected across randomizations
* $RFE$ (Recursive Feature Elimination): given an external estimator that assigns weights to features, the goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and weights are assigned to each one of them. Then, features whose absolute weights are the smallest are pruned from the current set features. That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached
* $RandomForestRegressor$: a random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and use averaging to improve the predictive accuracy and control over-fitting
* $ExtraTreesRegressor$: this class implements a meta estimator that fits a number of randomized decision trees (a.k.a. extra-trees) on various sub-samples of the dataset and use averaging to improve the predictive accuracy and control over-fitting
* $SelectKBest$: select features according to the k highest scores
* $SelectFpr$: FPR test stands for False Positive Rate test. It controls the total amount of false detections

```python
X = training_data(words_list3, words_list_trend, prec_week=False)
Y = gt_all
names = words_list3+words_list_trend 
 
ranks = {}

# function that create a dictionary of type {'Feature_selection_approach': {word1: rank, word2: rank...}}
def rank_to_dict(ranks, names, order=1):
    minmax = MinMaxScaler() # scale according min and max
    ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x, 2), ranks)
    return dict(zip(names, ranks ))

lr = LinearRegression(normalize=True)
lr.fit(X, Y)
ranks["Linear_reg"] = rank_to_dict(np.abs(lr.coef_), names)

ridge = Ridge(alpha=7)
ridge.fit(X, Y)
ranks["Ridge"] = rank_to_dict(np.abs(ridge.coef_), names)

lasso = Lasso()
lasso.fit(X, Y)
ranks["Lasso"] = rank_to_dict(np.abs(lasso.coef_), names)

rlasso = RandomizedLasso(alpha=0.04)
rlasso.fit(X, Y)
ranks["Stability"] = rank_to_dict(np.abs(rlasso.scores_), names)

rfe = RFE(lr, n_features_to_select=5)
rfe.fit(X,Y)
ranks["RFE"] = rank_to_dict(map(float, rfe.ranking_), names, order=-1)

rfr = RandomForestRegressor()
rfr.fit(X,Y)
ranks["RFR"] = rank_to_dict(rfr.feature_importances_, names)

exr = ExtraTreesRegressor()
exr.fit(X,Y)
ranks['EXR'] = rank_to_dict(exr.feature_importances_, names)

kbest = SelectKBest()
kbest.fit(X,Y)
ranks['KBEST'] = rank_to_dict(np.abs(kbest.scores_), names)

fpr = SelectFpr()
fpr.fit(X,Y)
ranks['FPR'] = rank_to_dict(np.abs(fpr.scores_), names)

# calculate means of all ranks
means = {} 
for name in names:
    means[name] = round(np.mean([ranks[method][name] 
                             for method in ranks.keys()]), 2)

ranks["Mean"] = means

table_rank = pd.DataFrame.from_dict(ranks)
cols_order = [ 'Lasso', 'Linear_reg', 'RFE', 'Ridge', 'Stability', 'RFR', 'EXR', 'KBEST','FPR','Mean']

table_rank = table_rank[cols_order].sort_values(by='Mean', ascending = False)
table_rank
```

```python
# function that fit the model with the firsts words in the mean ranking
# with_prec = Boolean, select or not the precending week values
def model_validation(list_words, list_trend, with_prec):
    train = training_data(list_words, list_trend, prec_week=with_prec)
    if with_prec == False:
        test = np.array(gt_all)
    else:
        test = np.array(gt_all[1:])
    
    # cross-validation
    scores = []
    for i in range(10):
        X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.3)
        linear_regr = sklearn.linear_model.LinearRegression()
        linear_regr_fit = linear_regr.fit(X_train, y_train)
        scores.append(linear_regr_fit.score(X_test, y_test))

    predicted = cross_val_predict(linear_regr, train, test, cv=10) # prediction
    rmse = statsmodels.tools.eval_measures.rmse(predicted, test) # rmse measure
    return [np.mean(scores), rmse, predicted]
```

```python
number_features = {} # dictionary of type {number of features: [mean scores, rmse]}
for i in range(1,9): # select a maximum of 8 features
    first_list = list(table_rank.iloc[range(i)].index.values) # first i words
    list_words = []
    list_trend = []
    for w in first_list:
        if w in words_list3:
            list_words.append(w)
        else:
            list_trend.append(w)
    number_features[i] = model_validation(list_words, list_trend, False)[0:2]
```
