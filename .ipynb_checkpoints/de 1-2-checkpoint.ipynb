```python
def csvToSeries(f,sep,dec):
    df = pd.read_csv(f,sep=sep,skiprows=[0,1],header=None,index_col=0,decimal=dec)
    s = df.iloc[:,0] #parse_dates=True, infer_datetime_format=False
    s = s.str.replace('.', '').apply(int)
    return s
```

def hourlyToWeek(f,s): #format 2016-53
    c=pd.read_csv(f, sep = s, header=None,parse_dates=True, infer_datetime_format=True,index_col = 0)[1:]
    c['Week'] = c.index.to_datetime()
    c = c.sort_index()
    c['Week'] = c['Week'].dt.year.astype(str) + '-' + c['Week'].dt.week.astype(str)
    c = c.reset_index(drop=True)
    t = c.groupby(c['Week'],sort=False)[1].sum()
    return t

```python
def hourlyToWeekly(f,s,idx): #format: 2016-01-03 BUT drop the index and keep only the values
    t = pd.Series.from_csv(f, sep = s, header=None, index_col = 0, parse_dates=True, infer_datetime_format=False)[1:]
    t = t.resample('W').sum()
    t = t.reset_index(drop=True)
    #newt = pd.concat([t,pd.Series(idx)], axis = 1,ignore_index=True) #if you want the ISO weeks as index
    return t
```

```python
#a function to drop night hours
def dropNight(f,s,beg,end):
    dt = pd.Series.from_csv(f,sep = s, header=None, parse_dates=[0], infer_datetime_format=False, index_col = 0)[1:]
    dt = dt.ix[dt.index.indexer_between_time(datetime.time(beg), datetime.time(end),
                                    include_start=False, include_end=False)]
    dt = dt.resample('W').sum()
    dt = dt.reset_index(drop=True)
    return dt
```
