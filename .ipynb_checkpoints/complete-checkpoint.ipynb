# Influnet time-series

```python
def csvToSeries(f,sep,dec):
    df = pd.read_csv(f,sep=sep,skiprows=[0,1],header=None,index_col=0,decimal=dec)
    s = df.iloc[:,3] 
    s = s.reset_index(drop=False)
    s.columns = ['date','v']
    s['date'] = s['date'].apply(lambda x: Week(int(x.split('-')[0]),int(x.split('-')[1])).monday())
    s = s.set_index(s['date'])
    s = s['v'].apply(float)
    return s
```

```python
dfInflu = pd.Series()
for f in glob.glob(subInflu + "/*.txt"):
    dfInflu = dfInflu.append(csvToSeries(f,' ',','))
```

```python
dfInflu.plot()
plt.show()
```

```python
def hourlyToWeekly(f,s):
    t = pd.Series.from_csv(f, sep = s, header=None, index_col = 0, parse_dates=True, infer_datetime_format=False)[1:]
    t = t.resample('W-MON').sum()
    t = t.dropna()
    return t
```

```python
#a function to drop night hours
def dropNight(f,s,beg,end):
    dt = pd.Series.from_csv(f,sep = s, header=None, parse_dates=[0], infer_datetime_format=False, index_col = 0)[1:]
    dt = dt.ix[dt.index.indexer_between_time(datetime.time(beg), datetime.time(end),
                                    include_start=False, include_end=False)]
    dt = dt.resample('W-MON').sum()
    dt = dt.dropna()
    return dt
```

# features selection

```python
names = ["x%s" % i for i in range(1,len(dfKeysDiv.columns)+1)] #number of variables to rank
methods = ['fReg','mutInf','Corr'] #methods to select top K features
ranks = {}

f ,pval = f_regression(dfKeysDiv, df['Ground'])
ranks["fReg"] = {names[k]:round(f[k],3) for k in range(len(f))}

mi = mutual_info_regression(dfKeysDiv, df['Ground'])
ranks['mutInf'] = {names[k]:round(mi[k],3) for k in range(len(mi))}

ranks['Corr'] ={names[k]:round(abs(groundCorr[k]),3) for k in range(len(names))}

r = {}
for name in names:
    r[name] = round(np.mean([ranks[method][name] 
                             for method in ranks.keys()]), 2)
    
ranks["Mean"] = r
methods.append("Mean")

i = 0
print "\t%s" % "\t".join(methods), '\tFeature'
for name in names:
    print "%s\t%s" % (name, "\t".join(map(str,[ranks[method][name] for method in methods]))), '\t' + dfKeysDiv.columns[i]    
    i += 1
```

```python
#top 5 features
count = 0
top5 = []
for name in names:
    if r[name] in sorted(r.values())[-5:]:
        print name,dfKeysDiv.columns[count]
        top5.append(dfKeysDiv.columns[count])
    count += 1
```

```python
kf = KFold(n_splits=30).split(dfKeysDiv[top5])
models = ['linearReg', 'ridgeReg', 'lassoReg', 'ranFor','regTree','bayReg','polyReg']
mseDict = {k:[] for k in models}
X = dfKeysDiv[top5]
Y = df['Ground']

for train,test in kf:
    
    X_train, X_test = X.ix[train,:], X.ix[test,:]
    y_train, y_test = Y.ix[train], Y.ix[test]
    
    #linear regression
    linearReg = linear_model.LinearRegression(fit_intercept=True, normalize=True).fit(X_train,y_train)
    LSyHat = linearReg.predict(X_test) 
    mseDict['linearReg'].append(metrics.mean_squared_error(y_test, LSyHat))
    
    #Ridge regression
    ridgeRegCv = linear_model.RidgeCV(alphas=[0.1, 1.0, 10.0]).fit(X_train,y_train)       
    a = ridgeRegCv.alpha_  
    ridgeReg = Ridge(alpha=a).fit(X_train,y_train)
    RidY = ridgeReg.predict(X_test)
    mseDict['ridgeReg'].append(metrics.mean_squared_error(y_test, RidY))
    
    #Lasso regression
    lassoReg = Lasso(alpha=.05).fit(X_train,y_train)
    lassY = lassoReg.predict(X_test)
    mseDict['lassoReg'].append(metrics.mean_squared_error(y_test, lassY))
    
    #Random Forest
    ranFor = RandomForestRegressor().fit(X_train,y_train)
    rfY = ranFor.predict(X_test)
    mseDict['ranFor'].append(metrics.mean_squared_error(y_test, rfY))
    
    #Regression Tree
    regTree = DecisionTreeRegressor(max_depth=3).fit(X_train,y_train)
    rtY = regTree.predict(X_test)
    mseDict['regTree'].append(metrics.mean_squared_error(y_test, rtY))
    
    #Bayesian regression
    bayReg = linear_model.BayesianRidge().fit(X_train,y_train)
    BayY = bayReg.predict(X_test)
    mseDict['bayReg'].append(metrics.mean_squared_error(y_test,BayY))
    
    #Polynomial regression
    polyReg = Pipeline([('poly', PolynomialFeatures(degree=4)),
                   ('linear', LinearRegression(fit_intercept=False, normalize=True))])
    polyReg = polyReg.fit(X_train,y_train)
    modY = polyReg.predict(X_test)
    mseDict['polyReg'].append(metrics.mean_squared_error(y_test,modY))   
```

```python
mseMean = {k:np.mean(mseDict[k]) for k in models}
mseMean
```

```python
minPair = min(mseMean.items(), key=lambda k: k[1])
print 'The model that gives minimum MSE is ', minPair[0], ' with ', minPair[1]
```

```python
ranFor = RandomForestRegressor(warm_start=True).fit(X,Y)
rfY = ranFor.predict(X)
print 'MSE using Random Forest Regressor: ',metrics.mean_squared_error(Y, rfY)
plt.plot(df['Ground'].index.to_pydatetime(),rfY, label="Yhat")
plt.plot(df['Ground'].index.to_pydatetime(),df['Ground'].values, label="Ground Truth")
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.legend()
plt.show()
```
